```python

import cv2
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img =  cv2.imread('img.png',0)
dim=(128,32)
nimg=cv2.resize(img,dim)
img =cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
ret,thresh1 = cv2.threshold(nimg,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh1)
# plt.imshow(pimg)
thresh1.shape
```

```python
data = bytearray(thresh1)
data
```

```python
import math
nd =list()    
with open("imgcode.txt","w")  as file:
    for i in range(0,4096,16):
        byte=0
        for j in range(0,16):
            byte+=data[i+j]
        st = "{0}, ".format(hex(math.ceil(byte/255)))
        file.write(st)
        if(i%15==0):
            file.write("\n")
    
```
